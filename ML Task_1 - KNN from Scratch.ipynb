### `Standard Scalar`

```python
X = data.drop('price',axis=1)
y = data['price']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['color'] = le.fit_transform(X['color'])
X['clarity'] = le.fit_transform(X['clarity'])
X['cut'] = le.fit_transform(X['cut'])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[['carat','table','x','y','z']] = sc.fit_transform(X[['carat','table','x','y','z']])
X
```

```python
#X = np.array(X)
X
X = X[:100]
y = y[:100]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=42)
```

## `KNN from scratch`

```python
def KNN(k,X_train,y_train,X_test):
    predicted = np.zeros(X_test.shape[0])                                               # store prediction
    for i in range(X_test.shape[0]):                                                    # test sample times
        d = np.zeros(X_train.shape[0])                                                  # store distance as train times
        for j in range(X_train.shape[0]):                                               # train sample times
            dist = np.sqrt(np.sum((X_train.values[j] - X_test.values[i])**2))           # calculating distance
            print(f'i => {i} and j => {j}')
            print(f'Distance => {dist}')
            d[j] = dist                                                                 # storing distance
            d = d.argsort()                                                             # sorting the arguments of distance
            print(f'Sorted Arguments => {d}')
            neighbor = y_train.values[d[:k]]                                            # finding neighbors
            print(f'All Neighbors sorted acc to distance => {y_train.values[d]}\nNearest Neighbor => {neighbor}')
            predicted[i] = np.mean(neighbor)                                            # doing the mean of neighbors
            print(f'Predicted =>{predicted[i]}\n\n\n\n')
    return predicted
```

```python
y_pred_scratch = KNN(3,X_train,y_train,X_test)
y_pred_scratch
```

## `KNN sklearn`

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=13)   #by-default n_neighbors=5
knr.fit(X_train,y_train)
y_pred = knr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(f'R2 => {r2_score(y_test,y_pred)}')
print(f'RMSE => {np.sqrt(mean_squared_error(y_test,y_pred))}')
```
