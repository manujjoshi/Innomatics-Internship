# **Predicting the consumption of electricity in the coming future using FBprophet**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('/content/Electric_Production.csv')
data.head()
```

```python
data.columns=['ds','y']
```

```python
data
```

```python
from fbprophet import Prophet
```

```python
px.line(x=data['ds'],y=data['y'])
```

```python
mdl = Prophet(seasonality_mode='multiplicative')
mdl.fit(data)
```

```python
future = mdl.make_future_dataframe(periods=36,freq='MS')
```

```python
forcasted = mdl.predict(future)
```

```python
forcasted
```

```python
data['y'].plot(figsize = (12, 5), legend = True)
forcasted["yhat"].plot(legend = True)
```

```python
fig2 = mdl.plot_components(forcasted)
```
