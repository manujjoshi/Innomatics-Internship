# Mean <font size="5">($\mu$):</font>
- Mean is a measure of central tendency which is the average of all the numbers given and is denoted by <font size="5">$\mu$</font> and is also called as Arithmetic mean
- Average is computed by adding all the numbers together and dividing then by number of elements.
- For e.g:
    - Consider following list of numbers is given <br>
    <b>2, 7, 4, 6, 3, 8, 5, 4, 9, 1, 2, 5, 2, 7, 5, 2, 8, 8, 3, 8</b>
    - To Compute the mean we will add the numbers <br>
    <b>2 +  7 +  4 +  6 +  3 +  8 +  5 +  4 +  9 +  1 +  2 +  5 +  2 +  7 +  5 +  2 +  8 +  8 +  3 +  8 $=$ <i>99</i></b>
    - We Will divide the sum of the aboove numbers by total number of elements i.e<i><b>20</b></i><br><br>
    <font size="4"><b>
    $\frac{2 +  7 +  4 +  6 +  3 +  8 +  5 +  4 +  9 +  1 +  2 +  5 +  2 +  7 +  5 +  2 +  8 +  8 +  3 +  8}{20} = $ $\frac{99}{20} \Rightarrow$4.95</b> </font>
- **Formula**:&nbsp;
<font size="6">$\mu=\frac{\Sigma X}{n}$</font>

# Median

- The median of a set of data is the middlemost number or center value in the set.Half of the data lies below and aboove 50th Percentile.
- To find the median, the data should be arranged, first, in order of least to greatest or greatest to the least value.
- For e.g:
    - Consider following list of numbers is given <br>
    <b>2, 7, 4, 6, 3, 8, 5, 4, 9, 1, 2, 5, 2, 7, 5, 2, 8, 8, 3, 8</b>

        - Sort the numbers in ascending order <br>

        - Find Middle most value<br>
            - Since we have even set of numbers, The Middle values will consist two values and we take its mean as median<br>

# Mode

- Mode is a value which occurs frequently in a set of values.
- There are two types of mode
    - **Unimodal**
        - It picks out the single frequent value occuring in dataset regardless how many other numbers frequently occur.
        - For e.g:<br>
            - <b>1, 2, 2, 2, 2, 3, 3, 4, 4, 5, 5, 5, 6, 7, 7, 8, 8, 8, 8,9</b><br>
            - In the aboove set 2 occurs frequently so as per unimodal value only 2 will be the mode.
    -**Multi-Modal**
        - Unlike Unimodal,Multi modal is opposite and gets more than one frequent values.
        - For e.g :<br>
            - <b>1, 2, 2, 2, 2, 3, 3, 4, 4, 5, 5, 5, 6, 7, 7, 8, 8, 8, 8,9</b><br>
            - In The aboove set along with 2,8 also has same number of counts,so the multi modes will be 2 and 8
- **Using Library function**:

```python
df['No_of_Fly_Members'].mode()
```

- **Without Using Library function**:

```python
## Mode means maximum frequency
def CountFrequency(my_list):
 
    # Creating an empty dictionary
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1

    keymax = max(zip(freq.values(), freq.keys()))[1]
    return (keymax)
```

# Variance ($\sigma^2$)
- Variance can be defined as a number which determines the spread from the mean. In simple terms how far the value is from mean on positive or negative side and is denoted by <b><font size="5"> $\sigma^2$</font></b>
- A large variance indicates that numbers in the set are far from the mean and far from each other. A small variance, on the other hand, indicates the opposite. A variance value of zero, though, indicates that all values within a set of numbers are identical.
- **Formula**:
    - <b><font size="5"> $\sigma^2$=$\frac{\sum\limits_{i=1}^n(x_i-\bar{x})^2}{n-1}$</font>
    - Where 
        - $x_i=$ <font size="3">$i^{th}$ point in dataset</font>
        - $\bar{x}=$ <font size="3">Mean of dataset</font>
        - n$=$ <font size="3">Number of elements in dataset</font>
- The aboove formula was for Sample,For Population too the formula is same but slight difference that instead of <b>n-1</b> we use <b>n</b>
- **Using Library function**:

```python
#with pacakge
df['Mthly_HH_Income'].var()
```

- **Without Using Library function**:

```python
sums = 0
for i in df['Mthly_HH_Income']:
    sums += (i - df['Mthly_HH_Income'].mean() ) ** 2
var = sums / (len(df['Mthly_HH_Income']) - 1)
var
```

# Standard Deviation($\sigma$)

- Standard Deviation is the measure of the dispersion of data from its mean. It measures the absolute variability of a distribution and is denoted by Symbol <font size="4">$\sigma$</font>
- The higher is the dispersion or variability of data, the larger will be the standard deviation and the larger will be the magnitude of the deviation of value from the mean whereas the lower is the dispersion or variability of data, the lower will be the standard deviation and the lower will be the magnitude of the deviation of value from the mean.
- The standard deviation formula is used to find the values of a specific data that is dispersed from the mean value. It is important to observe that the value of standard deviation can never be negative.
- Types:
    - Population
    - Sample
- **Formula**:
    - <font size="5">$\sigma$=$\sqrt{\frac{\sum\limits_{i=1}^n(x_i-\bar{x})^2}{n-1}}$</font>
        - Where 
            - $x_i=$ <font size="3">$i^{th}$ point in dataset</font>
            - $\bar{x}=$ <font size="3">Mean of dataset</font>
            - n$=$ <font size="3">Number of elements in dataset</font>
- The aboove formula was for Sample,For Population too the formula is same but slight difference that instead of <b>n-1</b> we use <b>n</b>
- **Using Library function**:

```python
#with pacakge
df['Mthly_HH_Income'].std()
```

- **Without Using Library function**:

```python
def std(data):
    sums = 0
    for i in data:
        sums += (i - data.mean() ) ** 2
    var = sums / (len(data) - 1)
    return (np.sqrt(var))
std(df['Mthly_HH_Income'])
```

# Correlation(r)
- Correlation also called as pearson correlation refers to a process for establishing the relationships between two variables.
- The correlation coefficient, r, is a summary measure that describes the extent of the statistical relationship between two interval or ratio level variables. The correlation coefficient is scaled so that it is always between -1 and +1. When r is close to 0 this means that there is little relationship between the variables and the farther away from 0 r is, in either the positive or negative direction, the greater the relationship between the two variables.
- Types Of Correlation:
    - Strong Or Positive Correlation
    - Weak Or Negative Correlation
    - No correlation
- **Formula**:
    - <font size="5">$r = \frac{n\sum{XY}-(\sum{X}\sum{Y})}{\sqrt{ [n \sum{x^2}-(\sum{x})^2 ][n \sum{y^2}-(\sum{y})^2 }]} $</font><br>
        - Where
            - n=Number of elements
            - $\sum{XY}$= Sum of Product of X and Y
            - $\sum{X}$= Sum of X
            - $\sum{Y}$= Sum of Y

# QQ Plot
- The Q-Q plot, or quantile-quantile plot, is a graphical tool to help us assess if a set of data plausibly came from some theoretical distribution such as a Normal or exponential. For example, if we run a statistical analysis that assumes our dependent variable is Normally distributed, we can use a Normal Q-Q plot to check that assumption. Its just a visual check, not an air-tight proof, so it is somewhat subjective. But it allows us to see at-a-glance if our assumption is plausible, and if not, how the assumption is violated and what data points contribute to the violation.
- A Q-Q plot is a scatterplot created by plotting two sets of quantiles against one another. If both sets of quantiles came from the same distribution, we should see the points forming a line thats roughly straight. Heres an example of a Normal Q-Q plot when both sets of quantiles truly come from Normal distributions.

- Since The aboove Plot Goes More in linear way we can say the distribution is Normal

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
data = np.random.normal(-5,1, 10)
fig = sm.qqplot(data, line='45')
plt.show()
```

# Box Cox
- When the assumption of data normally distributed is violated or the relationship between the dependent and independent variables in case of linear model are not linear, in such situations some transformations methods that may help the data set follow a normal distribution. Box Cox is one such transformation method.
- The basic assumption of Box-Cox is data must be positive (no negative values) and also data should be continuous.
- The assumption of any linear modes is that relationship between the response variable Y and the predictor variable Xs is linear, however this is the not true the all the times, so when the relationship between the dependent variable and independent variable is not linear and still wish to fit a linear model to the data then consider a Box-Cox transformation method. This will transform the predictor variable or the response variable and then fit a linear model to the data to study the effect that the predictor variable has on the transformed responses.


```python
#Before COX transformation
import numpy as np 
from scipy.stats import boxcox 
import seaborn as sns 
np.random.seed(0)
data = np.random.exponential(size=5000)
sns.distplot(data, kde=True,color="purple") 
```

```python
#After Cox transformation
transformed_data, best_lambda = boxcox(data) 
sns.distplot(transformed_data, kde=True,color="purple") 
```
